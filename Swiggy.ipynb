```python
df_Swiggy["Cuisine"].unique()
```

```python
freq_dict={}
for i in df_Swiggy["Cuisine"].unique():
    cuisine_list=i.split(",")
    for x in cuisine_list:
        x=x.lstrip(" ")
        if x in freq_dict:
            freq_dict[x]=freq_dict[x]+1
        else:
            freq_dict[x]=1
print(freq_dict)
```

```python
cuisine=freq_dict.keys()
print(cuisine)
count=freq_dict.values()
print(count)
```

```python
cuisine_Swiggy=pd.DataFrame({"Cuisine":cuisine,"Count":count})
```

```python
cuisine_Swiggy
```

```python
plt.figure(figsize=(17,8))
sns.barplot(y="Count",x="Cuisine",data=cuisine_Swiggy,palette="husl")
plt.xticks(rotation=90)
plt.show()
```

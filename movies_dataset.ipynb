The Best and worst movies
in terms of budget,revenue,profit,ratings....


```python
from IPython.display import HTML
```

```python
df_best=df[['poster_path','title','budget_musd','revenue_musd','vote_count','vote_average','popularity']].copy()
```

```python
df_best.head()
```

```python
df_best['profit_musd']=df_best['budget_musd'].sub(df['revenue_musd'])
```

```python
df_best.head()
```

```python
df_best['return']=df_best['revenue_musd'].div(df['budget_musd'])
df_best
```

```python
df_best.columns=['','Title','Budget','Revenue','Votes','Average Rating','Popularity','Profit','ROI']
df_best
```

```python
df_best.set_index('Title',inplace=True)
```

```python
df_best
```

```python
subset=df_best.iloc[:5, :2]
subset
```

```python
HTML(subset.to_html(escape=False))
```

```python
df_best.sort_values(by='Average Rating',ascending=True)
```

```python
df_best.sort_values(by='ROI',ascending=False)
```

```python
df_best.loc[df_best['Budget']>=5].sort_values(by='ROI',ascending=False)
```

```python
df_best['Budget'].fillna(0,inplace=True)
df_best['Votes'].fillna(0,inplace=True)
```

```python
df_best['Votes'].isnull().sum()
```

```python
df_best.info()
```

```python
def best_worst(n,by,ascending=False,min_bud=0,min_votes=0):
    df2=df_best.loc[(df_best['Budget']>=min_bud)&(df_best['Votes']>=min_votes),['',by]].sort_values(by=by,ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

```python
best_worst(5,'Revenue')
```

```python
best_worst(2,'Budget')
```

```python
best_worst(5,'Profit')
```

```python
best_worst(5,'Profit',ascending=True)
```

```python
best_worst(5,'ROI',min_bud=10)
```

```python
best_worst(5,'ROI',ascending=True,min_bud=100)
```

```python
best_worst(5,'Votes')
```

```python
best_worst(5,'Average Rating',min_votes=1000)
```

Top movies sorted based lowest rating and minimum votes of x


```python
best_worst(5,'Average Rating',ascending=True,min_votes=1000)
```

```python
#Top movies sorted based lowest rating and minimum votes of x and minimum budget of y
```

```python
best_worst(5,'Average Rating',ascending=True,min_bud=10,min_votes=10000)
```

```python
#top movies sorted on popularity
best_worst(5,'Popularity')
```
